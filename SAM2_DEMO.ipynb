```python
import base64

def encode_image(filepath):
    with open(filepath, 'rb') as f:
        image_bytes = f.read()
    encoded = str(base64.b64encode(image_bytes), 'utf-8')
    return "data:image/jpg;base64,"+ encoded

def encode_image2(img, format='.jpg'):
    """
    Encodes an OpenCV image (NumPy array) to a base64 string.

    Parameters:
    - img: OpenCV image (NumPy array).
    - format: Image format for encoding (default is '.jpg').

    Returns:
    - A base64-encoded string representing the image.
    """
    success, buffer = cv2.imencode(format, img)
    if not success:
        raise ValueError("Image encoding failed.")
    encoded = base64.b64encode(buffer).decode('utf-8')
    return f"data:image/{format.lstrip('.')};base64,{encoded}"


IS_COLAB = True

if IS_COLAB:
    from google.colab import output
    output.enable_custom_widget_manager()

from jupyter_bbox_widget import BBoxWidget

widget = BBoxWidget()
widget.image = encode_image2(frame) #encode_image("/content/TESTRENDER.PNG")
widget
```

```python
selected_points =[[dict_["x"],dict_["y"]] for dict_ in widget.bboxes]
```

```python
from ultralytics.models.sam import SAM2VideoPredictor

# Create SAM2VideoPredictor
overrides = dict(conf=0.25, task="segment", mode="predict", imgsz=1024, model="sam2_b.pt")
predictor = SAM2VideoPredictor(overrides=overrides)

# Run inference with selected points
results = predictor(source=video_path, points=selected_points, labels=[1]*len(selected_points))
```

```python
# Open the original video to get frame properties
import numpy as np
import os

cap = cv2.VideoCapture(video_path)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
cap.release()

# Define the codec and create VideoWriter object
out_path = os.path.basename(video_path).split(".")[0] + "_masked.mp4"

out = cv2.VideoWriter(out_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))


# Process each result frame
for result in results:
    # Assuming result.masks.data is a list of masks for the frame
    if result.masks:
        # Select the main mask (e.g., the one with the highest confidence)

        mask_data = result.masks.data.cpu().numpy()  # Shape: (N, H, W)

        # Select the first mask (or iterate over all masks as needed)
        mask = mask_data[0]  # Shape: (H, W)

          # Convert mask to 8-bit unsigned integer format
        mask = (mask * 255).astype(np.uint8)
        # main_mask = result.masks[0]  # Adjust index if needed

        # # Convert mask to uint8

        # Create a 3-channel mask image
        mask_rgb = cv2.merge([mask, mask, mask])

        # Optionally, overlay the mask on a black background
        masked_frame = cv2.bitwise_and(mask_rgb, mask_rgb)

        # Write the frame to the output video
        out.write(masked_frame)

# Release the VideoWriter
out.release()
```

```python
# Open the original video to get frame properties
cap = cv2.VideoCapture(video_path)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Read all frames into a list
frames = []
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    frames.append(frame)
cap.release()

# Define the codec and create VideoWriter object
out_path_colored = os.path.basename(video_path).split(".")[0] + "_masked_colored.mp4"
out = cv2.VideoWriter(out_path_colored, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

# Process each result and frame
for idx, result in enumerate(results):
    if result.masks and idx < len(frames):
        frame = frames[idx]

        # Get mask data
        mask_data = result.masks.data.cpu().numpy()  # Shape: (N, H, W)
        mask = mask_data[0]  # Assuming the first mask

        # Convert mask to 8-bit
        mask = (mask * 255).astype(np.uint8)

        # Create a color version of the mask
        color = (0, 255, 0)  # Green mask for example
        mask_color = np.zeros_like(frame)
        for c in range(3):  # Apply the color to the mask
            mask_color[:, :, c] = np.where(mask > 0, color[c], 0)

        # Blend the mask with the frame
        alpha = 0.5  # Transparency factor
        blended = cv2.addWeighted(frame, 1 - alpha, mask_color, alpha, 0)

        # Write the blended frame to output
        out.write(blended)
    else:
        # If no mask, write the original frame
        if idx < len(frames):
            out.write(frames[idx])

# Release the VideoWriter
out.release()
```
